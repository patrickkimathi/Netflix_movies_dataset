```python
#format the columns to be consistent 
def clean_columns(netflix_df):
    """
    remove all the whitespaces
    convert the columns to lowercase
    the white spaces in between the words in column with an underscore"""
    netflix_df.columns =   netflix_df.columns.str.strip().str.lower().str.replace(' ', '-')
    return netflix_df
    
#call the clean_cuolumns() function
netflix_df = clean_columns(netflix_df)

#display the the cleaned columns
print(netflix_df.columns)
```

# ** Handling Missing values for Director column**

```python
#list of the Director Cast pairs and the number of times they apear
netflix_df['dir_cast'] = netflix_df['director']+'---'+ netflix_df['cast']
```

```python
netflix_df['dir_cast']
```

```python
#count the uniques values  in the dir_cast pair
counts = netflix_df['dir_cast'].value_counts()
counts
```

```python
#Check the dir_cast pair with more than or equal 3 counts
filtered_counts = counts[counts>=3]
filtered_counts 
```

```python
#Get the values 
filtered_values = filtered_counts.index
```

```python
#convert the values into a list
list_dr_cast =list(filtered_values)

#declare a dictionary
dict_direcast = dict()
#
for i in list_dr_cast:
    director, cast  = i.split('---')
    dict_direcast[director] = cast
```

```python
#Fill in the missing directors
for i in range(len(dict_direcast)):
    netflix_df.loc[(netflix_df['director'].isna()) & (netflix_df['cast'] ==
    list(dict_direcast.items())[i][1]), 'director'] =list(dict_direcast.items())[i][0]
```

```python
netflix_df.isnull().sum()
```

```python
#Assign 'NotGiven to all other null valuea not filled 
netflix_df.loc[netflix_df['director'].isna(),'director'] = 'Not_Given'
```

```python
#then confirm the missing values 
netflix_df.isnull().sum()
# Missing values in the Director column successfully filled
```

# Missing Values in the Country Column

****Fill in the missing countries********
#Given that there 831 missing countries
#We could use information from the directors to fill in missing countries


```python
#Use directors to fill in the missing countries 
directors = netflix_df['director']
countries = netflix_df['country']
```

```python
#Pair each director with their country use zip() to get an iterator tuples
pairs = zip(directors , countries)
```

```python
#Convert the list of tuples inot a dictionary
dir_cntry = dict(list(pairs))
```

```python
#Director matched to country values used to fill in null country values
for i in range(len(dir_cntry)):
    netflix_df.loc[(netflix_df['country'].isna())&(netflix_df['director'] == 
                                                list(dir_cntry.items())[i][0]),'country']=list(dir_cntry.items())[i][1]
```

```python
netflix_df.isnull().sum()
```

97 missing countries imputed
